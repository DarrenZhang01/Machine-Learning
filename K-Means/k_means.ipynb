```python
# The practice code is from "https://cloud.tencent.com/developer/article/1085115"
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
X, y = make_blobs(n_samples=1000, n_features=2, centers=[[-1, -1], [0, 0], [1, 1], [2, 2]],
                 cluster_std=[0.1, 0.2, 0.3, 0.2], random_state = 10)
plt.scatter(X[:, 0], X[:, 1], marker = 'o')
plt.show()
```

```python
y_pred = KMeans(n_clusters = 2, random_state = 10).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
metrics.calinski_harabaz_score(X, y_pred)
```

```python
y_pred = KMeans(n_clusters = 3, random_state = 10).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
metrics.calinski_harabaz_score(X, y_pred)
```
